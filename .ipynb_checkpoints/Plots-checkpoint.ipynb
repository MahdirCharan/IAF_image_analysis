In these set of experiments Iodixanol **10%** in the center and Iodixano **36%** in the side inlet were used. Single frequency and low applied voltage were impelementd.

```python
ax = sns.violinplot(x="Qcenter = 7.5 μl/min", y="impedance", hue="prefocusing",
                    data=result, palette="muted", split=True, inner="quartile")
ax.set_title("Distribution of impedance for Jurkat cells");
```

### Cell's location along the width of the channel while prefocusing is on
- **Kernel density estimation**: A histogram aims to approximate the underlying probability density function that generated the data by binning and counting observations. Kernel density estimation (KDE) presents a different solution to the same problem. Rather than using discrete bins, a KDE plot <u>smooths the observations with a Gaussian kernel</u>, producing a continuous density estimate.

```python
data = result[result["prefocusing"]=="Yes"]
hist = sns.displot(data, x="cols", hue="Qcenter = 7.5 μl/min",kind="kde", multiple="stack")
hist.set(xlabel="Width of the channel");
```
